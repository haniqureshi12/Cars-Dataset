visualizing the missing/null values count

```python
# Calculate missing values count per column
missing_values = df.isnull().sum()
missing_values[missing_values > 0].sort_values(ascending=False).plot(
    kind='bar', figsize=(12, 6), color='teal')

plt.title("Count of Missing Values Per Column", fontsize=16)
plt.xlabel("Columns", fontsize=14)
plt.ylabel("Count of Missing Values", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout for better display
plt.show()
```

DATA CLEANING

```python
# droping a unwnted column
df = df.drop('Max Torque', axis=1)
```

HANDLING THE MISSING VALUES
one by one column

```python
df['Fuel Tank Capacity'] = df['Fuel Tank Capacity'].fillna(df['Fuel Tank Capacity'].mean())
```

```python
df['Engine'] = df['Engine'].fillna(df['Engine'].mode()[0])
```

```python
df['Length'] = df['Length'].fillna(df['Length'].median()) 
```

Fill missing values Dynamically

```python
# Fill missing values dynamically based on data type
for column in df.columns:
    if df[column].dtype in ['int64', 'float64']:  # Numerical columns
        if df[column].skew() > 1:  # Skewness > 1 indicates outliers
            df[column] = df[column].fillna(df[column].median())  # Use Median
        else:
            df[column] = df[column].fillna(df[column].mean())  # Use Mean
    elif df[column].dtype == 'object':  # Categorical columns
        df[column] = df[column].fillna(df[column].mode()[0])  # Use Mode
```

NOW DATA IS CLEAN 

```python
df.isnull().sum()
```

* Define numerical & categorical colunms

```python
# define numerical & categorical colunms
numeric_features = [feature for feature in df.columns if np.issubdtype(df[feature].dtype, np.number)]
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'object']

#print colunms
print('We have {} numerical features:{}'.format(len(numeric_features),numeric_features))
print('\nWe have {} categorical features:{}'.format(len(categorical_features),categorical_features))
```

Features Info

```python
for col in categorical_features:
    print(df[col].value_counts(normalize=True) * 100)
    print('----------------------------')
```

Univariate Analysis

The term univariate refers to the analysis of one variable prefix "uni" mean "one". The  puprose of univariate analysis is to understand the distribution of values for a single variable.

```python
# Set the figure size and title
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8,y=1.02)
# Loop through numerical features and plot each one
for i in range(0,len(numeric_features)):
    plt.subplot(3,3,i+1)
    sns.kdeplot(x=df[numeric_features[i]],shade=True,color='g')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

Categorical Feature

```python
# Define categorical columns
import math
cat1 = ['Make', 'Model', 'Fuel Type', 'Transmission', 'Location', 
        'Color', 'Owner', 'Seller Type', 'Engine', 'Max Power', 'Drivetrain']
# Calculate rows and columns dynamically
num_features = len(cat1)
rows = math.ceil(num_features / 2)  # 2 columns per row

# Set figure size
plt.figure(figsize=(15, 5 * rows))
plt.suptitle('Univariate Analysis of Categorical Features', 
             fontsize=20, fontweight='bold', alpha=0.8, y=1.02)

# Loop through categorical features and plot each one
for i, feature in enumerate(cat1):
    plt.subplot(rows, 2, i + 1)
    sns.countplot(x=df[feature], palette='viridis')  # Add a color palette for better visuals
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.tight_layout()

# Show the plots
plt.show()
```
